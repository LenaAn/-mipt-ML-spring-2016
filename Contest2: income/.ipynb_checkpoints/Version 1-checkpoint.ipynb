```python
gain_minus_loss = (data[:, 11] - data[:, 12]) * 1.
gain_minus_loss_test = (test_data[:, 11] - test_data[:, 12]) * 1.
```

```python
fnlwgt_strng = (data[:, 3] > 800000) * 1.0
fnlwgt_strng_test = (test_data[:, 3] > 800000) * 1.0

for i in range(len(data[:, 3])):
    if data[i, 3] > 800000:
        data[i, 3] = 800000
        
for i in range(len(test_data[:, 3])):
    if test_data[i, 3] > 800000:
        test_data[i, 3] = 800000
```

```python
capital_gain_strange = (data[:, 11] > 40000) * 1.0
capital_gain_strange_test = (test_data[:, 11] > 40000) * 1.0

for i in range(len(data[:, 1])):
    if data[i, 11] > 40000:
        data[i, 11] = 40000

for i in range(len(test_data[:, 1])):
    if test_data[i, 11] > 40000:
        test_data[i, 11] = 40000
```

```python
data = np.column_stack((data, gain_minus_loss, fnlwgt_strng, 
                        capital_gain_strange))
test_data = np.column_stack((test_data, gain_minus_loss_test, fnlwgt_strng_test, 
                        capital_gain_strange_test))
```

Развернем текстовые признаки

```python
text_feature_indexs = [2, 4, 6, 7, 8, 9, 10, 14]
```

```python
def categorial_string_to_numbers(values):
    uniq_values = set(values)
    values_to_num = dict(zip(values, range(len(values))[::-1]))
    return map(lambda x: values_to_num[x], values)
```

```python
for i in text_feature_indexs:
    data[:, i] = list(categorial_string_to_numbers(data[:, i]))
    
for i in text_feature_indexs:
    test_data[:, i] = list(categorial_string_to_numbers(test_data[:, i]))
```

```python
test_data
```

```python
data_ohe = OneHotEncoder(categorical_features=text_feature_indexs).fit_transform(data)
test_data_ohe = OneHotEncoder(categorical_features=text_feature_indexs).fit_transform(test_data)
```
