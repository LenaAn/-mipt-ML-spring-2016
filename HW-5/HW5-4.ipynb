```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
grid = np.mgrid[-2:2:0.05, -2:2:0.05]
func = (grid[0] - 1)**2 + 2*(grid[0]**2 - grid[1] - 1)**2

plt.figure(figsize=(8, 8))
CS = plt.contour(grid[0], grid[1], func, [0.01, 0.05, 0.2, 0.5, 1, 2, 3, 4, 6, 8, 10])  #    
plt.clabel(CS, fontsize=10, inline=1, fmt='%1.3f')
plt.xlim((np.min(grid[0]), np.max(grid[0])))
plt.ylim((np.min(grid[1]), np.max(grid[1])))
    
plt.show()
```
